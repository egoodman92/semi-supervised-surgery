```python
!pwd
from csv import reader
import csv
import cv2
from pathlib import Path


label_to_class = {"bovie":0, "forceps":1, "needledriver":2, "hand":3}


def scale_and_transform_annotation(image):    


    #extract picture name
    image_name = image[0][0][-25:]
    print("Image name is", image_name)
    
    image_loc = "../../images/" + image_name
    new_loc = "./"
    !cp $image_loc $new_loc
    
    if image[0][2] == "":
        with open(image_name[:-4]+'.txt', 'w') as fp:
            pass    

    else:
        #convert "needledriver" to 1"
        for label in image:
            label[5] = label_to_class[label[5]]



        img = cv2.imread(image_name)
        dimensions = img.shape
        print("\n\nWorking with image", image_name)
        print("Image dimensions are", dimensions)

        new_labels = []
        for label in image:
            og = [float(x) for x in label[1:]]
            print("Original annotations are", og)
            final_spatial = [int(label[5]), \
                             (og[2]+og[0]) / 2 / dimensions[1], \
                             (og[3]+og[1]) / 2 / dimensions[0], \
                             (og[2]-og[0]) / dimensions[1], \
                             (og[3]-og[1]) / dimensions[0]
                             ]
            new_labels.append(final_spatial)

        print("Final spatial annotations are", new_labels)

        print("Exporting new annotation to", image_name[:-4]+".txt")
        with open(image_name[:-4]+'.txt', 'w') as f:
            wr = csv.writer(f,delimiter=" ")
            wr.writerows(new_labels)
    
    
with open('test_data.csv') as read_obj:
    csv_reader = reader(read_obj)
    list_of_images = list(csv_reader)

past_image = list_of_images[0][0]
cur_accumulation = []
for i, image in enumerate(list_of_images):
    #print("Studying image", i)
    if image[0] == past_image:
        cur_accumulation.append(image)
    else:
        #print("\n\nSCALING", cur_accumulation)
        scale_and_transform_annotation(cur_accumulation)
        past_image = image[0]
        cur_accumulation = [image]

        
print("\n\nFINALSCALING", cur_accumulation)
scale_and_transform_annotation(cur_accumulation)

    
```
